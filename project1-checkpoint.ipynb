```python
import numpy as np
```

```python
c=0
D = 20
Foodnum = 20
lb = -100
ub = 100
limit = 100
maxcycle = 3000
Food = np.random.randint(lb,ub,size = (Foodnum,D))
#Food matrix is a matrix which consists of a randomly initialised solution matrix
Food1 = np.random.randint(lb,ub,size = (Foodnum,D))
#Food1 matrix is the shortest position value matrix of Food matrix
fitness1 = np.zeros(shape = (Foodnum,1))
fit = np.zeros(shape = (Foodnum,D))
fitness = np.zeros(shape = (Foodnum,1))
prob = np.zeros(shape = (Foodnum,1))
fooddup = np.zeros(shape = (Foodnum,D))
trial = np.zeros(shape = (Foodnum,1))
print(Food)
```

```python
city = np.random.randint(1,100,size = (D,D))
for i in range(D):
    for j in range(D):
        if(i==j):
            city[i][j] = 0
        else:
            city[j][i] = city[i][j]
#city is a matrix which stores the distances between two cities where rows and column numbers represent cities
print(city)
```

```python
def SPV(Food,index):
    Fooddup = np.copy(Food)
    c = 0
    for j in range(D):
        minindex = 0
        min = Food[index][0]
        for i in range(D):
            if(Food[index][i] < min):
                min = Food[index][i]
                minindex = i
        Food1[index][minindex] = c
        Food[index][minindex] = 100000000
        c+=1
    Food[index] = np.copy(Fooddup[index])
```

```python
def SPV1(solution):
    soldup = np.copy(solution)
#     print(solution)
#     print(soldup)
    c=0
    solution1 = np.zeros(shape = (D,))
    for j in range(D):
        minindex=0
        min=solution[0]
        for i in range(D):
            if(solution[i]<min):
                min=solution[i]
                minindex=i
        solution1[minindex]=c
        solution[minindex]=10000000
        c+=1
#     print(soldup)
#     print(solution)
    solution = np.copy(soldup)
    return solution,solution1
```

```python
def CalculateFitness(Food1,index):
    Food1.astype(int)
    sum = 0
    for j in range(D-1):
        sum += city[Food1[index][j]][Food1[index][j+1]]
    sum += city[Food1[index][D-1]][Food1[index][0]]
    return sum
```

```python
def CalculateFitness1(sol):
    sol = sol.astype(int)
    sum = 0
    for i in range(D-1):
        sum += city[sol[i]][sol[i+1]]
    sum += city[sol[D-1]][sol[0]]
    return sum
```

```python
def membestsource():
    globalmin = np.amin(fitness)
    bestsol = np.argmin(fitness)
    return globalmin,bestsol
```

```python
def Empbee(Food,Food1,fitness,trial):
    for i in range(Foodnum):
        solution = np.zeros(shape = (D,))
        solution1 = np.zeros(shape = (D,))
        param2change = np.random.randint(0,D)
        neighbour = np.random.randint(0,Foodnum)
        while(neighbour==i):
            neighbour = np.random.randint(0,Foodnum)
        for j in range(D):
            solution[j] = Food[i][j]
        r = np.random.uniform(0.0,1.0)
        solution[param2change]=Food[i][param2change]+(Food[i][param2change]-Food[neighbour][param2change])*(r-0.5)*2
#         print(Food[i][param2change])
#         print(solution[param2change])
        if (solution[param2change]<lb):
             solution[param2change]=lb
        if (solution[param2change]>ub):
            solution[param2change]=ub
        solution,solution1 = SPV1(solution)
        Fitnesssol = CalculateFitness1(solution1)
        
        if(Fitnesssol < fitness[i]):
            #print("yes")
            trial[i] = 0
            Food1[i][j] = solution1[j]
            fitness[i] = Fitnesssol
            for j in range(D):
                Food[i][j] = solution[j]
           # print(Food[i])
        else:
            trial[i] += 1
```

```python
def Calculateprob(fitness):
    sum = np.sum(fitness)
    for i in range(Foodnum):
        prob[i] = fitness[i]/sum
```

```python
 def Onlookbee(Food,Food1,fitness,trial):
        solution = np.zeros(shape = (D,))
        solution1 = np.zeros(shape = (D,))
        t = 0
        i = 0
        while(t < Foodnum):
            r = np.random.uniform(0,1)
            if(r<prob[i]):
                t +=1
                param2change = np.random.randint(0,D)
                neighbour = np.random.randint(0,Foodnum)
                while(neighbour==i):
                    neighbour = np.random.randint(0,Foodnum)
                for j in range(D):
                    solution[j] = Food[i][j]
                r = np.random.uniform(0.0,1.0)
                solution[param2change]=Food[i][param2change]+(Food[i][param2change]-Food[neighbour][param2change])*(r-0.5)*2
#         print(Food[i][param2change])
#         print(solution[param2change])
                if (solution[param2change]<lb):
                     solution[param2change]=lb
                if (solution[param2change]>ub):
                     solution[param2change]=ub
                solution,solution1 = SPV1(solution)
                Fitnesssol = CalculateFitness1(solution1)
        
                if(Fitnesssol < fitness[i]):
                    #print("yes")
                    trial[i] = 0
                    Food1[i][j] = solution1[j]
                    fitness[i] = Fitnesssol
                    for j in range(D):
                        Food[i][j] = solution[j]
           # print(Food[i])
                else:
                    trial[i] += 1
            i += 1
            if(i == Foodnum-1):
                i = 0
```

```python
def Scoutbee(Food,fitness,trial):
    for i in range(Foodnum):
        if(trial[i] >= limit):
            for j in range(D):
                Food[i][j] = np.random.randint(lb,ub)
            SPV(Food,i)
            fitness[i] = CalculateFitness(Food1,i)
            trial[i] = 0
```
